```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../experiments/results.csv")
df.head()
```

```python
df.columns
```

```python
if "epoch" not in df.columns:
    df["epoch"] = df.index

df["epoch"] = df["epoch"].astype(int)
df.head()
```

```python
box_loss_col = None
cls_loss_col = None

for col in df.columns:
    if "box" in col and "loss" in col:
        box_loss_col = col
    if "cls" in col and "loss" in col:
        cls_loss_col = col

print("Box loss column:", box_loss_col)
print("Class loss column:", cls_loss_col)

# Plot
plt.figure(figsize=(10,5))
plt.plot(df["epoch"], df[box_loss_col], label="Box Loss")
plt.plot(df["epoch"], df[cls_loss_col], label="Class Loss")

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Curves")
plt.legend()
plt.show()
```

```python
precision_col = None
recall_col = None

for col in df.columns:
    if "precision" in col.lower():
        precision_col = col
    if "recall" in col.lower():
        recall_col = col

print("Precision Column:", precision_col)
print("Recall Column:", recall_col)

plt.figure(figsize=(10,5))
plt.plot(df["epoch"], df[precision_col], label="Precision")
plt.plot(df["epoch"], df[recall_col], label="Recall")

plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Precision & Recall Curve")
plt.legend()
plt.show()
```

#  Interpretation of Model Training Results

- **Box Loss** decreased  model learned to localize objects.
- **Class Loss** decreased  model improved classification.
- **Precision is high (~70%+)**  few false positives.
- **Recall is low (~35%)**  model misses many real objects.
- **mAP is moderate (~40%+)**  model performance is acceptable but can be improved.

##  Why recall is low:
- Small dataset
- Low-quality images
- Poor lighting conditions
- Not enough variation
- Need augmentation
